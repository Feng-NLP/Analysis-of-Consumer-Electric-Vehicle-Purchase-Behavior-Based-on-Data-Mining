```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV,cross_val_score,KFold,train_test_split
from sklearn import  metrics
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score,accuracy_score
from sklearn.metrics import cohen_kappa_score

for i in range(5,16):
#训练集与测试集划分
    X_res_1 = X_res.iloc[:,0:i]
    x_train, x_test, y_train, y_test = train_test_split(X_res_1, y_res, test_size=0.3,random_state=7)
#构建并训练logistics模型
    clf2 = XGBClassifier(use_label_encoder=False,objective='binary:logistic', eval_metric='error')
    clf2.fit(x_train,y_train)

#将训练好的模型用于测试集的预测
    y_pred_prob2 = clf2.predict_proba(x_test)
    y_pred_prob2 = pd.DataFrame(y_pred_prob2)
    y_pred2 = clf2.predict(x_test)
    y_true = y_test
#保存测试集预测概率，用于绘制ROC曲线
#y_pred_prob.to_csv('Logistic测试集预测概率.csv')

#混淆矩阵
#classes = list(set(y_true))
#classes.sort()
#confusion = confusion_matrix(y_pred2, y_true)
#plt.imshow(confusion, cmap=plt.cm.Blues)
#indices = range(len(confusion))
#plt.xticks(indices, classes)
#plt.yticks(indices, classes)
#plt.colorbar()
#plt.xlabel('y_pred')
#plt.ylabel('y_true')
#for first_index in range(len(confusion)):
#    for second_index in range(len(confusion[first_index])):
 #       plt.text(first_index, second_index, confusion[first_index][second_index])
# plt.show()

    print(i)
#kappa系数
    kappa = cohen_kappa_score(y_true,y_pred2)
    print("kappa:",kappa)

#正确率
    print("ACC：",accuracy_score(y_true, y_pred2))

#精确率
    print("precision：",precision_score(y_true, y_pred2))

#F1分数
    print("F1_score",f1_score(y_true, y_pred2))    
```
